```python
import os
import numpy as np
import cv2
from torchvision.transforms import transforms 
import torch
import torchvision
import json
```

```python
train_path = "./trimmed/val/"
inf_path = "./val_inf/"

cls = os.listdir(train_path)
cls
```

```python

model = torchvision.models.detection.keypointrcnn_resnet50_fpn(pretrained=True,
                                                               num_keypoints=17)
# set the computation device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# load the modle on to the computation device and set to eval mode
model.to(device).eval()
transform = transforms.Compose([
    transforms.ToTensor()
])
```

```python
from PIL import Image
from tqdm import tqdm



with torch.no_grad():
    for i, name in tqdm(enumerate(cls)):
        for cl in tqdm(os.listdir(os.path.join(train_path,name))):
            
            video_path = os.path.join(os.path.join(train_path,name),cl)
            
            if str(cl)+'_data_val.json' in os.listdir(inf_path):
#                 print(f"Skipping {cl}")
                continue
            
            landmarks_ls = [] 

            cap = cv2.VideoCapture(video_path)
            wd = 0

            while(cap.isOpened()):
            # Capture frame-by-frame
                ret, frame = cap.read()
                
                
                if ret:
                    wd+=1
                    
                    if wd % 5 == 0:
                        try:

                            PIL_image = Image.fromarray(frame.astype('uint8'), 'RGB')

                            image = transform(PIL_image)
                            image = image.unsqueeze(0).to(device)

                            outputs = model(image)
#                             idx = np.where(outputs[0]["scores"].cpu().max())
                            pts = outputs[0]["keypoints"][0].cpu().numpy()
                            landmarks_ls.append(np.squeeze(pts).tolist())

                        except:
                            print(f"Some error while inferencing in {cl}")
                            print(outputs[0]["scores"].cpu())
                    
                    
                else:
                    vid_data = {}
                    vid_data["filename"] = {"name":cl,
                                            'class':i ,
                                            'keypoints':landmarks_ls} 

                    break

            
#             data["video_key_point"].append(vid_data)

            with open(os.path.join(inf_path,str(cl)+'_data_val.json'), 'w') as f:
                json.dump(vid_data, f)
     
                    
```
